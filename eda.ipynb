## Data Wrangling

Vamos a normalizar los datos y los vamos a unir en uno solo para trabajarlos de forma más cómoda

```python
df_2021 = pd.read_csv('data/delitos_2021.csv', sep=';')
df_2020 = pd.read_csv('data/delitos_2020.csv')
df_2019 = pd.read_csv('data/delitos_2019.csv')
df_2018 = pd.read_csv('data/delitos_2018.csv')
df_2017 = pd.read_csv('data/delitos_2017.csv')
```

Creamos una lista que contenga todos los datasets para trabajar de forma cómoda

```python
lista_df = [df_2021,df_2020,df_2019,df_2018,df_2017]
```

Verificamos las columnas de todas los df

```python
for i in lista_df:
    print(i.columns, len(i.columns))
```

Los **datasets de 2017 a 2019 tienen las mismas columnas**, a partir del 2020 hay variaciones tanto con los anteriores como con los posteriores.

id-mapa, id_mapa e id refieren todos al mismo dato -> **id**

franja y franja horaria remiten al mismo dato -> **franja_horaria**

tipo y tipo delito refieren a los mismos datos - **tipo_delito**

victimas, cantidad_registrada y cantidad refieren a los mismos datos -> **cantidad_registrada**

subtipo y subtipo_delito -> **subtipo_delito**

latitud y longitud -> **lat** y **long**

anio-mes-día viene en la columna **fecha**, nos quedamos con esta ultima en principio

Vamos a utiliar las columnas de los dataframes viejo.


```python
df_2021.columns = ['id', 'anio', 'mes', 'dia', 'fecha', 'franja_horaria', 'tipo_delito', 'subtipo_delito',
'uso_armas', 'barrio', 'comuna', 'lat', 'long', 'cantidad_registrada']

df_2020.columns = ['id', 'anio', 'mes', 'dia', 'fecha', 'franja_horaria', 'tipo_delito', 'subtipo_delito',
'uso_armas', 'barrio', 'comuna', 'lat', 'long', 'cantidad_registrada']

df_2021.drop(['anio', 'mes', 'dia'],inplace=True,axis=1)

df_2020.drop(['anio', 'mes', 'dia'],inplace=True,axis=1)
```

Parseamos todas las columnas de fechas y rezamos con que infer datetime funcione :)

```python
for i in lista_df:
    i.fecha = pd.to_datetime(i.fecha, infer_datetime_format=True)
```

```python
df = pd.concat(lista_df, axis=0).drop('id', axis=1)
```

```python
df
```

```python
df.info()
```

Las variables con gran cantidad de nulos son uso_armas y subtipo_delito. Verificamos a que tipo de delitos corresponden los nulos.

```python
print(f"El total de entradas de tipo de delito para las entradas con uso de arma nulos: {df[df.uso_armas.isna()].tipo_delito.value_counts().sum()}")
print('\nSu distribución es:')
df[df.uso_armas.isna()].tipo_delito.value_counts(normalize=True)
```

La mayoría corresponde a sustracciones de elementos de valor en sus tipos sin violencia y con violencia, este último al no reportarse armas podría corresponder a violencia de tipo física. Por último las lesiones y homicidios pueden corresponder a delitos usando violencia física.

```python
print(f"El total de entradas de subtipo de delito para las entradas con uso de arma nulos: {df[df.uso_armas.isna()].subtipo_delito.value_counts().sum()}")
print('\nSu distribución es:')
df[df.uso_armas.isna()].subtipo_delito.value_counts(normalize=True)
```

Analizamos los subtipos de delitos en caso de existir (recordemos que contiene muchos nulos también):

```python
df[df.uso_armas.isna()].pivot_table(
    index=['tipo_delito','subtipo_delito'],
    values='fecha',
    aggfunc='count'
)/df[df.uso_armas.isna()].pivot_table(
    index=['tipo_delito','subtipo_delito'],
    values='fecha',
    aggfunc='count'
).sum()*100
```

Observamos que en la mayoría de los casos son hurtos y lesiones, los cuales involucran automoviles, con lo cual que no involucre armas o que al menos no se hayan reportado tiene sentido con lo cual se rellenara con "No aplica". En el caso de los robos tanto con auto como con moto también es lógico que existieran robos sin uso de arma o que al menos no se haya registrado. Por último en el caso de los homicidios pueden corresponder a homicidios usando violencia física o en situaciones viales en el subtipo que le corresponde.

```python
df[df.subtipo_delito.isna()].tipo_delito.value_counts()
```

En el caso las entradas con NaN en subdelito, analizando la distribución de los delitos asociados podemos decir que probablemente correspondan a robos y hurtos simples y quizá peleas callejeras en el caso de las lesiones. Con lo cuál se rellenaran con "No aplica".

```python
df[df.cantidad_registrada.isna()].tipo_delito.value_counts()
```

En el caso de la cantidad de victimas registradas puede que no se hayan cargado, con lo cuál se rellenara como desconocido.

```python
df.subtipo_delito.fillna('No aplica', inplace=True)
df.uso_armas.fillna('No aplica', inplace=True)
df.cantidad_registrada.fillna('Desconocido', inplace=True)
```

Por último revisamos como queda el dataframe.

```python
df.info()
```

Por último se dropean los NaN que quedan ya que son una cantidad despreciable para el tamaño del dataset

```python
df.dropna(inplace=True)
df.reset_index(inplace=True)
```

Ya tenemos nuestro dataset limpio!! :D

## EDA

### ¿Como ha evolucionado la cantidad de delitos desde el 2017 al 2021?

Podemos empezar por ver que fechas son las que registran más registros, pero quizá sea más informativo analizar a través de los gráficos

```python
df.fecha.value_counts()
```

Buscamos el top 10 de los barrios con más delitos usando value_counts()

```python
df.barrio.value_counts().head(10)
```

Usamos boolean indexing para quedarnos con los registros que apuntan a los 10 barrios con más delitos

```python
df_top_10 = df[df.barrio.isin(df.barrio.value_counts().head(10).index)]
```

Mapeamos los colores de los tres barrios más habitados según los datos de wikipedia

```python
colores = df_top_10.barrio.value_counts().index.sort_values().map({
    'Almagro':'slateblue',
    'Balvanera':'sandybrown',
    'Barracas':'lightgrey',
    'Belgrano':'lightgrey',
    'Caballito':'lightgrey',
    'Flores':'lightgrey',
    'Palermo':'lightgrey',
    'Recoleta':'lightcoral',
    'San Nicolás':'lightgrey',
    'Villa Lugano':'lightgrey'
})
```

Realizamos un groupby por barrio y fecha con la cantidad de entradas y usamos unstack para quitar el multiindex

```python
df_top_10.groupby(['fecha','barrio']).size().unstack()
```

Graficámos nuestros datos con plotly y le ponemos los colores mapeados

```python
fig = px.line(df_top_10.groupby(['fecha','barrio']).size().unstack(), color_discrete_sequence=colores,height=900)
fig.update_layout(
        title=dict(text='Cantidad de crimenes anuales por barrio (Top 10)',x=0.53,y=0.95,
                   font=dict(size=20)),
        xaxis_title_text='Año',
        xaxis = dict(
                tickmode = 'array',
                tickvals = [2017,2018,2019,2020,2021]),
        yaxis_title_text='Conteo',
        legend=dict(title='Barrio',x=1.05,y=0.96,borderwidth=0,tracegroupgap=5),
        )
```

Como se puede ver es bastante ruidoso plotear día a día, podemos visualizar usando una agrupación por año en vez de por día

```python
fig = px.line(df_top_10.groupby([df.fecha.dt.year,'barrio']).size().unstack(), color_discrete_sequence=colores)

fig.update_layout(
        title=dict(text='Cantidad de crimenes anuales por barrio (Top 10)',x=0.53,y=0.95,
                   font=dict(size=20)),
        xaxis_title_text='Año',
        xaxis = dict(
                tickmode = 'array',
                tickvals = [2017,2018,2019,2020,2021]),
        yaxis_title_text='Conteo',
        legend=dict(title='Barrio',x=1.05,y=0.96,borderwidth=0,tracegroupgap=5),
        )
```

```python
df_mensual_anual_10 = df_top_10.set_index('fecha').groupby([pd.Grouper(freq="M"), 'barrio']).size().unstack()
```
